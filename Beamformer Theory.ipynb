```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 64
pos = 500*np.random.uniform(size=N)-250
plt.scatter(pos, np.ones(64))
plt.show()
```

```python
c = 3e8
N_BEAMS = 256+1
N_FREQ = 256
theta = np.linspace(0, np.pi, N_BEAMS)
freq = np.linspace(1.28,1.53, N_FREQ)
A = np.zeros((N_BEAMS, N), dtype=np.complex64)
# for k in range(N_FREQ):
k = 0
wavelength = c/(freq[k]*1e9)
for i in range(N_BEAMS):
    for j in range(N):
        A[i,j] = np.exp(-2.0j*np.pi*pos[j]*np.cos(theta[i])/wavelength)
```

```python
signal = [np.exp(2*np.pi*1j*pos[i]*np.cos(theta[200]+.1)/wavelength) for i in range(N)]
out = np.dot(A, signal)
plt.scatter(theta, np.real(out))
```
