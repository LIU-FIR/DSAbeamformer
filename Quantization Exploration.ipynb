```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1000000
t = np.linspace(0, 2*np.pi,N)
x_quant = np.around(127*np.cos(t))
y_quant = np.around(127*np.sin(t))
```

```python
mag_quant = np.sqrt(x_quant*x_quant + y_quant*y_quant)
phase_quant = np.arctan2(y_quant, x_quant)
```

```python
mag_error = mag_quant/127
phase_error = t - phase_quant
wrapped_phase_error = [phase_error[i] if phase_error[i] < np.pi else phase_error[i]-np.pi*2 for i in range(len(phase_error))]
```

```python
plt.hist(mag_error, bins=30)
plt.title("Fourier Coefficient Quantization Amplitude Error (N={})".format(N))
plt.ylabel("N")
plt.xlabel("Amplitude Error")
plt.show()

plt.hist(wrapped_phase_error)
plt.title("Fourier Coefficient Quantization Phase Error (N={})".format(N))
plt.ylabel("N")
plt.xlabel("Phase Error [rad]")
plt.show()
```

```python
fig = plt.figure(figsize=(7,7))
fig.add_subplot(111,aspect='equal')
plt.hist2d(np.cos(t)-x_quant/127, np.sin(t)-y_quant/127, bins = 300)
plt.xlabel("x-axis error")
plt.ylabel("y-axis error")
plt.title("Histogram of Fourier Coefficient Errors")
plt.savefig("Histogram.png")
plt.show()
```
